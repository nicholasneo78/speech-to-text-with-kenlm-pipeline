```python
# get samples of the main data from the pkl format
import numpy as np
import pandas as pd
from tqdm import tqdm
import pickle
```

```python
DATA_PATH_TRAIN = './root/pkl/librispeech_train.pkl'
DATA_PATH_DEV = './root/pkl/librispeech_dev.pkl'
DATA_PATH_TEST = './root/pkl/librispeech_test.pkl'

# open and load the train dataset
with open(DATA_PATH_TRAIN, 'rb') as f:
    df_train = pickle.load(f)

df_train
```

```python
# sample the data
df_train_sample = df_train.sample(frac=0.1, random_state=42)
df_train_sample
```

```python
# place your data entries into a DataFrame (for multiple dataframe)
combined_data_train = pd.DataFrame()
combined_data_train = combined_data_train.append(df_train_sample)
combined_data_train = combined_data_train.sample(frac=1, random_state=42) # randomise the data if there are multiple dataset used
combined_data_train
```

```python
# once the training data is sampled merge back with the dev and the test data

# open and load the dev and test dataset
with open(DATA_PATH_DEV, 'rb') as f:
    df_dev = pickle.load(f)
    
with open(DATA_PATH_TEST, 'rb') as f:
    df_test = pickle.load(f)
```

```python
# place your data entries into a DataFrame (for multiple dataframe)
combined_data_dev = pd.DataFrame()
combined_data_dev = combined_data_dev.append(df_dev)
combined_data_dev = combined_data_dev.sample(frac=1, random_state=42) # randomise the data if there are multiple dataset used
combined_data_dev
```

```python
combined_data_test = pd.DataFrame()
combined_data_test = combined_data_test.append(df_test)
combined_data_test = combined_data_test.sample(frac=1, random_state=42) # randomise the data if there are multiple dataset used
combined_data_test
```

```python
combined_data_train.to_pickle()
combined_data_dev.to_pickle()
combined_data_test.to_pickle()
```
